```python
# Importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

# Load the dataset
data = pd.read_csv('canada_per_capita_income.csv')

# Inspect the first few rows
data.head()

# Plotting the data to see the relationship
# The column name in the dataframe is likely 'per capita income (US$)'
# Instead of 'Per Capita Income'. Changing it to the correct name.
plt.scatter(data['year'], data['per capita income (US$)'])
plt.xlabel('year')
plt.ylabel('Per Capita Income (in USD)')
plt.title('Canada Per Capita Income over the Years')
plt.show()

# Train/Test Split (Here, we are just using all data for prediction)
X = data['year'].values.reshape(-1, 1)  # Feature - Year
# Also changing the column name here to match the dataframe
y = data['per capita income (US$)'].values    # Target - Per Capita Income

# Initialize the Linear Regression model
model = LinearRegression()

# Train the model
model.fit(X, y)

# Predict for the year 2020
prediction_2020 = model.predict([[2020]])

print(f"Predicted Per Capita Income for Canada in 2020: {prediction_2020[0]}")
```

```python
# Importing required libraries
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer # Importing SimpleImputer

# Load the dataset
salary_data = pd.read_csv('salary.csv')

# Inspect the first few rows
salary_data.head()

# Train the model with the dataset
X = salary_data['YearsExperience'].values.reshape(-1, 1)  # Feature - Years of Experience
y = salary_data['Salary'].values  # Target - Salary

# Initialize the imputer to replace NaN with the mean
imputer = SimpleImputer(strategy='mean')

# Fit and transform the imputer on the feature data
X = imputer.fit_transform(X) # Replacing NaN values in 'X'

# Initialize the Linear Regression model
model = LinearRegression()

# Train the model
model.fit(X, y)

# Predict salary for 12 years of experience
predicted_salary = model.predict([[12]])

print(f"Predicted Salary for an employee with 12 years of experience: {predicted_salary[0]}")
```
